Tensors
=======

Tensors are a specialized data structure that are very similar to arrays
and matrices. In PyTorch, we use tensors to encode the inputs and
outputs of a model, as well as the model's parameters.

Tensors are similar to NumPy's ndarrays, except that tensors can run on
GPUs or other specialized hardware to accelerate computing. If you're
familiar with ndarrays, you'll be right at home with the Tensor API. If
not, follow along in this quick API walkthrough.


```python
import torch
import numpy as np
```

Tensor Initialization
=====================

Tensors can be initialized in various ways. Take a look at the following
examples:

**Directly from data**

Tensors can be created directly from data. The data type is
automatically inferred.


```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
```

**From a NumPy array**

Tensors can be created from NumPy arrays (and vice versa - see
`bridge-to-np-label`{.interpreted-text role="ref"}).


```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

**From another tensor:**

The new tensor retains the properties (shape, datatype) of the argument
tensor, unless explicitly overridden.
### 1. Συναρτήσεις _like (Το «Καλούπι»)

Οι συναρτήσεις που τελειώνουν σε `_like` χρησιμοποιούν έναν ήδη υπάρχοντα Tensor (π.χ. τον `x_data`) ως «καλούπι». Αντιγράφουν το σχήμα του (δηλαδή τις διαστάσεις του, π.χ. 2x2), χωρίς να χρειάζεται να το ορίσεις εσύ.

* **`torch.ones_like(x_data)`**: Δημιουργεί έναν νέο Tensor που είναι γεμάτος με άσσους (1). Αντιγράφει πιστά και τον τύπο δεδομένων (π.χ. ακέραιους).
* **`torch.rand_like(x_data, dtype=torch.float)`**: Δημιουργεί έναν νέο Tensor γεμάτο με τυχαίους δεκαδικούς αριθμούς (από το 0 έως το 1).
  * *Προσοχή:* Το `dtype=torch.float` αλλάζει ρητά τον τύπο δεδομένων σε δεκαδικούς, το οποίο είναι υποχρεωτικό εδώ γιατί η `rand` βγάζει μόνο δεκαδικούς.


```python
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")
```

**With random or constant values:**

`shape` is a tuple of tensor dimensions. In the functions below, it
determines the dimensionality of the output tensor.


### 2. Δημιουργία Tensors με καθορισμένο Σχήμα (Shape)

Σε αυτή την περίπτωση δημιουργούμε Tensors από το μηδέν, ορίζοντας εμείς τις διαστάσεις τους μέσω μιας πλειάδας (tuple) που συνήθως την ονομάζουμε `shape`. Π.χ., το `shape = (2, 3,)` υποδηλώνει πίνακα με 2 γραμμές και 3 στήλες.

* **`torch.rand(shape)`**: Νέος Tensor γεμάτος με τυχαίους δεκαδικούς αριθμούς.
* **`torch.ones(shape)`**: Νέος Tensor γεμάτος με άσσους (1).
* **`torch.zeros(shape)`**: Νέος Tensor γεμάτος με μηδενικά (0).

```python
shape = (2, 3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

**Standard numpy-like indexing and slicing:**
### 3. Slicing (Τεμαχισμός Tensors)

Μπορούμε να επιλέξουμε και να τροποποιήσουμε συγκεκριμένες γραμμές ή στήλες ενός Tensor χρησιμοποιώντας αγκύλες `[γραμμές, στήλες]`.

* **Παράδειγμα:** `tensor[:, 1] = 0`
  * Η άνω και κάτω τελεία (`:`) σημαίνει «όλες οι γραμμές».
  * Το `1` σημαίνει «η δεύτερη στήλη» (η αρίθμηση ξεκινά από το 0).
  * Αποτέλεσμα: Μηδενίζει ολόκληρη τη δεύτερη στήλη του πίνακα!

```python
tensor = torch.ones(4, 4)
tensor[:,1] = 0
print(tensor)
```

**Joining tensors** You can use `torch.cat` to concatenate a sequence of
tensors along a given dimension. See also
[torch.stack](https://pytorch.org/docs/stable/generated/torch.stack.html),
another tensor joining op that is subtly different from `torch.cat`.


### 4. Ένωση Tensors: torch.cat vs torch.stack

* **`torch.cat` (Το σελοτέιπ):** Κολλάει τους Tensors τον έναν δίπλα στον άλλον (`dim=1`) ή τον έναν κάτω από τον άλλον (`dim=0`). Μεγαλώνει τον πίνακα, **χωρίς** να προσθέτει νέα διάσταση.
* **`torch.stack` (Η στοίβα):** Βάζει τον έναν Tensor πάνω στον άλλον (σαν σελίδες χαρτιού). Δημιουργεί πάντα μια **νέα διάσταση** (π.χ. από 2D πίνακες φτιάχνει έναν 3D).

```python
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

**Multiplying tensors**
### 5. Κατά στοιχείο πολλαπλασιασμός (Element-wise product)

Πολλαπλασιάζει τα στοιχεία των Tensors που βρίσκονται στην **ίδια ακριβώς θέση**.
Στο PyTorch, γράφεται με δύο ισοδύναμους τρόπους:
* Με συνάρτηση: `tensor.mul(tensor)`
* Με το σύμβολο: `tensor * tensor` (Πιο σύνηθες)

```python
# This computes the element-wise product
print(f"tensor.mul(tensor) \n {tensor.mul(tensor)} \n")
# Alternative syntax:
print(f"tensor * tensor \n {tensor * tensor}")
```

This computes the matrix multiplication between two tensors
### 6. Πολλαπλασιασμός Πινάκων (Matrix Multiplication)

Εκτελεί τον κλασικό γραμμικό πολλαπλασιασμό πινάκων (dot product), όπου οι γραμμές του πρώτου πολλαπλασιάζονται με τις στήλες του δεύτερου.
* **`tensor.T`**: Επιστρέφει τον Ανάστροφο (Transpose) του πίνακα (οι γραμμές γίνονται στήλες) για να ταιριάξουν οι διαστάσεις.
* **Τρόποι γραφής:** `tensor.matmul(tensor.T)` ή πιο απλά **`tensor @ tensor.T`**.

```python
print(f"tensor.matmul(tensor.T) \n {tensor.matmul(tensor.T)} \n")
# Alternative syntax:
print(f"tensor @ tensor.T \n {tensor @ tensor.T}")
```

**In-place operations** Operations that have a `_` suffix are in-place.
For example: `x.copy_(y)`, `x.t_()`, will change `x`.
### 7. In-place Πράξεις & Memory Sharing

* **In-place πράξεις (`_`):** Οποιαδήποτε συνάρτηση έχει μια κάτω παύλα `_` στο τέλος της (π.χ. `tensor.add_(5)`), τροποποιεί απευθείας τον ίδιο τον Tensor χωρίς να δημιουργεί αντίγραφο στη μνήμη. Το αντίστοιχο στο NumPy είναι το όρισμα `out=` (π.χ. `np.add(n, 1, out=n)`).
* **Memory Sharing:** Όταν μετατρέπουμε ένα NumPy array σε PyTorch Tensor στην CPU (με `torch.from_numpy()`), μοιράζονται την **ίδια ακριβώς μνήμη**. Αν αλλάξεις το ένα, αλλάζει αυτόματα και το άλλο!

```python
print(tensor, "\n")
tensor.add_(5)
print(tensor)
```

Bridge with NumPy {#bridge-to-np-label}
=================

Tensors on the CPU and NumPy arrays can share their underlying memory
locations, and changing one will change the other.


Tensor to NumPy array
=====================


```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

A change in the tensor reflects in the NumPy array.


```python
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

NumPy array to Tensor
=====================


```python
n = np.ones(5)
t = torch.from_numpy(n)
```

Changes in the NumPy array reflects in the tensor.


```python
np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
